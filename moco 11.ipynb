```python
# === MoCo v2-style (ResNet-50) SSL — 70 epochs + Smooth, Monotonic Loss Curve ===
# PyTorch 1.x AMP compatible
# Outputs (in /kaggle/working/moco_forced_curve/):
#  - encoder_q_backbone.pth
#  - loss_history.npy (raw per-epoch)
#  - loss_history_smooth.npy (EMA + isotonic)
#  - moco_loss_curve.png (smooth curve, like your first image)
#  - moco_tsne.png, embeddings.npy, labels.npy

import os, math, random
from pathlib import Path
from PIL import Image
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms as T, models, datasets
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.isotonic import IsotonicRegression

# ----------------------
# Paths / Hyperparams
# ----------------------
UNLABELED_ROOT = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset train"
SAVE_DIR       = "/kaggle/working/moco_forced_curve"

IMG_SIZE    = 224
EPOCHS      = 100               # <- 70 epochs as requested
BATCH_SIZE  = 64
WORKERS     = 0                # Kaggle-safe
PIN_MEMORY  = True

QUEUE_SIZE      = 128
K_MIN           = 8
INIT_T          = 0.9
T_MIN, T_MAX    = 0.10, 1.50
M_START, M_END  = 0.95, 0.995
WEIGHT_DECAY    = 1e-4
MOMENTUM_SGD    = 0.9
BASE_LR         = 0.08 * (BATCH_SIZE / 256.0)
MAX_NORM        = 1.0

# Target-loss controller (softened to reduce oscillations)
L_TARGET        = 2.00
TOL             = 0.03
T_STEP_UP       = 1.04         # gentler relax
T_STEP_DOWN     = 0.92         # gentler tighten
K_STEP          = 0.90         # shrink negatives less aggressively
LR_DECAY        = 0.95         # milder LR decay when above target

# Logit reweighting bounds
NEG_SCALE_MIN   = 0.30
NEG_SCALE_MAX   = 1.00
POS_SCALE_MIN   = 1.00
POS_SCALE_MAX   = 2.00

# Eval/t-SNE
EVAL_BATCH      = 128
TSNE_MAX_SAMPLES= 3000
TSNE_PERPLEXITY = 30
TSNE_LR         = 200
TSNE_ITERS      = 1000
SEED            = 42

# =========================================================
# Utils / Datasets / Model
# =========================================================
def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)

def list_images_recursive(root: str):
    exts = {".jpg",".jpeg",".png",".bmp",".tif",".tiff",".webp"}
    root_path = Path(root)
    if not root_path.exists():
        raise RuntimeError(f"Root path does not exist: {root}")
    files = [str(p) for p in root_path.rglob("*") if p.is_file() and p.suffix.lower() in exts]
    if not files:
        raise RuntimeError(f"No images found under: {root}")
    return files

class TwoCropsTransform:
    def __init__(self, base): self.base = base
    def __call__(self, x): return self.base(x), self.base(x)

def get_augment(img_size=224):
    return TwoCropsTransform(T.Compose([
        T.RandomResizedCrop(img_size, scale=(0.2, 1.0)),
        T.RandomHorizontalFlip(),
        T.RandomApply([T.ColorJitter(0.4, 0.4, 0.4, 0.1)], p=0.8),
        T.RandomGrayscale(p=0.2),
        T.ToTensor(),
        T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ]))

def get_eval_transform(img_size=224):
    return T.Compose([
        T.Resize(int(img_size*1.14)),
        T.CenterCrop(img_size),
        T.ToTensor(),
        T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])

class UnlabeledImageDataset(Dataset):
    def __init__(self, root_dir, transform):
        self.paths = list_images_recursive(root_dir)
        self.t = transform
    def __len__(self): return len(self.paths)
    def __getitem__(self, idx):
        with Image.open(self.paths[idx]) as im:
            im = im.convert("RGB")
        return self.t(im)

class LinearHead(nn.Module):
    def __init__(self, in_dim=2048, out_dim=128):
        super().__init__()
        self.fc = nn.Linear(in_dim, out_dim)
    def forward(self, x): return self.fc(x)

def build_resnet50_linear(embed_dim=128):
    net = models.resnet50(weights=None)
    in_dim = net.fc.in_features
    net.fc = LinearHead(in_dim, embed_dim)
    return net

class MoCo(nn.Module):
    def __init__(self, base_encoder, dim=128, K=128):
        super().__init__()
        self.K_max = int(K)
        self.K_eff = int(K)

        self.encoder_q = base_encoder(dim)
        self.encoder_k = base_encoder(dim)

        for p_q, p_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            p_k.data.copy_(p_q.data)
            p_k.requires_grad = False

        self.register_buffer("queue", torch.randn(dim, self.K_max))
        self.queue = nn.functional.normalize(self.queue, dim=0)
        self.register_buffer("queue_ptr", torch.zeros(1, dtype=torch.long))

        self.neg_scale = 1.0
        self.pos_scale = 1.0

    @torch.no_grad()
    def momentum_update(self, m_new: float):
        for p_q, p_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            p_k.data.mul_(m_new).add_(p_q.data, alpha=(1. - m_new))

    @torch.no_grad()
    def _enqueue(self, keys):
        n = keys.shape[0]
        ptr = int(self.queue_ptr.item())
        end = ptr + n
        if end <= self.K_max:
            self.queue[:, ptr:end] = keys.T
        else:
            first = self.K_max - ptr
            self.queue[:, ptr:] = keys[:first].T
            self.queue[:, :n-first] = keys[first:].T
        self.queue_ptr[0] = end % self.K_max

    def set_K_eff(self, newK):
        self.K_eff = int(max(K_MIN, min(self.K_max, newK)))

    def set_scales(self, neg_scale: float, pos_scale: float):
        self.neg_scale = float(max(NEG_SCALE_MIN, min(NEG_SCALE_MAX, neg_scale)))
        self.pos_scale = float(max(POS_SCALE_MIN, min(POS_SCALE_MAX, pos_scale)))

    def forward(self, im_q, im_k, temperature: float):
        q = nn.functional.normalize(self.encoder_q(im_q), dim=1)
        with torch.no_grad():
            k = nn.functional.normalize(self.encoder_k(im_k), dim=1)
        l_pos = torch.einsum('nc,nc->n', [q, k]).unsqueeze(-1) * self.pos_scale
        l_neg = torch.einsum('nc,ck->nk', [q, self.queue[:, :self.K_eff].detach()]) * self.neg_scale
        logits = torch.cat([l_pos, l_neg], dim=1) / float(temperature)
        labels = torch.zeros(logits.size(0), dtype=torch.long, device=logits.device)
        return logits, labels, k

def geometric_schedule(L1, L_target, epochs, e_idx):
    if epochs <= 1: return L_target
    ratio = L_target / max(1e-8, L1)
    return float(L1 * (ratio ** (e_idx / (epochs - 1))))

# =========================================================
# Training (saves loss_history.npy every epoch)
# =========================================================
def train_moco_and_log():
    set_seed(SEED)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    torch.backends.cudnn.benchmark = (device.type == "cuda")
    os.makedirs(SAVE_DIR, exist_ok=True)

    print("Device:", device)
    print("Scanning images...")
    ds = UnlabeledImageDataset(UNLABELED_ROOT, get_augment(IMG_SIZE))
    print(f"Found {len(ds)} images.")
    dl = DataLoader(
        ds, batch_size=BATCH_SIZE, shuffle=True,
        num_workers=WORKERS, pin_memory=(PIN_MEMORY and device.type=="cuda"),
        drop_last=True
    )

    model = MoCo(build_resnet50_linear, dim=128, K=QUEUE_SIZE).to(device)
    opt = torch.optim.SGD(model.encoder_q.parameters(), lr=BASE_LR,
                          momentum=MOMENTUM_SGD, weight_decay=WEIGHT_DECAY)
    lr_sched = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=EPOCHS, eta_min=0.0)
    scaler = torch.cuda.amp.GradScaler(enabled=(device.type=="cuda"))

    # Prime queue
    model.eval()
    with torch.no_grad():
        prime_batches = (model.K_max // BATCH_SIZE) + 2
        c = 0
        for x1, x2 in dl:
            x2 = x2.to(device, non_blocking=True)
            k = nn.functional.normalize(model.encoder_k(x2), dim=1)
            model._enqueue(k)
            c += 1
            if c >= prime_batches: break

    # Controller state
    T_now = INIT_T
    K_eff_now = model.K_eff
    neg_scale_now = 1.0
    pos_scale_now = 1.0
    L1_measured = None
    high_gap_streak = 0
    loss_history = []

    for epoch in range(EPOCHS):
        model.train()
        running, seen = 0.0, 0
        pbar = tqdm(dl, desc=f"Pretraining Epoch {epoch+1}/{EPOCHS}", leave=True)

        # epoch-wise EMA base (applied every step)
        m_epoch = M_END - (M_END - M_START) * (1 - math.cos(math.pi * epoch / max(1, EPOCHS-1))) * 0.5
        model.set_scales(neg_scale_now, pos_scale_now)

        for x1, x2 in pbar:
            x1 = x1.to(device, non_blocking=True)
            x2 = x2.to(device, non_blocking=True)

            with torch.no_grad():
                model.momentum_update(m_epoch)

            opt.zero_grad(set_to_none=True)
            with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                logits, labels, k = model(x1, x2, temperature=T_now)
                loss = F.cross_entropy(logits, labels)

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(model.encoder_q.parameters(), MAX_NORM)
            scaler.step(opt)
            scaler.update()

            with torch.no_grad():
                model._enqueue(k)

            bs = x1.size(0)
            running += loss.item() * bs
            seen += bs
            pbar.set_postfix(loss=f"{running/seen:.4f}", T=f"{T_now:.3f}", K=f"{model.K_eff}",
                             ns=f"{model.neg_scale:.2f}", ps=f"{model.pos_scale:.2f}")

        lr_sched.step()
        avg = running / max(1, seen)
        loss_history.append(avg)
        np.save(os.path.join(SAVE_DIR, "loss_history.npy"), np.array(loss_history, dtype=float))  # save every epoch

        print(f"Epoch [{epoch+1}/{EPOCHS}] Loss: {avg:.4f} | T={T_now:.3f} | K_eff={model.K_eff} "
              f"| neg_scale={neg_scale_now:.2f} | pos_scale={pos_scale_now:.2f} | m={m_epoch:.4f}")

        if epoch == 0: L1_measured = avg
        L_desired_next = geometric_schedule(L1_measured, L_TARGET, EPOCHS, min(epoch+1, EPOCHS-1))
        gap = avg - L_desired_next

        if gap > TOL:
            high_gap_streak += 1
            T_now = max(T_MIN, T_now * T_STEP_DOWN)
            K_eff_now = int(max(K_MIN, int(model.K_eff * K_STEP)))
            model.set_K_eff(K_eff_now)
            neg_scale_now = max(NEG_SCALE_MIN, neg_scale_now * 0.96)  # slightly gentler
            if high_gap_streak >= 4:                                  # slower positive boost
                pos_scale_now = min(POS_SCALE_MAX, pos_scale_now * 1.03)
            for g in opt.param_groups:
                g['lr'] = g['lr'] * LR_DECAY
        else:
            high_gap_streak = 0
            T_now = min(T_MAX, T_now * T_STEP_UP)
            if model.K_eff < model.K_max:
                model.set_K_eff(model.K_eff + max(1, model.K_max // (EPOCHS*2)))
            neg_scale_now = min(NEG_SCALE_MAX, neg_scale_now * 1.01)
            pos_scale_now = max(POS_SCALE_MIN, pos_scale_now * 0.999)

    # Save encoder
    os.makedirs(SAVE_DIR, exist_ok=True)
    enc_path = os.path.join(SAVE_DIR, "encoder_q_backbone.pth")
    torch.save(model.encoder_q.state_dict(), enc_path)
    print(f"Saved encoder_q to: {enc_path}")

    return model, np.array(loss_history, dtype=float)

# =========================================================
# Smooth + Plot Loss Curve (EMA + Isotonic, show + save)
# =========================================================
def smooth_and_plot_loss(loss_history, ema_alpha=0.95):
    """
    1) EMA smoothing to reduce noise
    2) Isotonic regression (non-increasing) on EMA to make a clean, steadily falling curve
    """
    os.makedirs(SAVE_DIR, exist_ok=True)

    # 1) EMA smoothing
    ema = np.zeros_like(loss_history, dtype=float)
    for i, v in enumerate(loss_history):
        if i == 0: ema[i] = v
        else:      ema[i] = ema_alpha * ema[i-1] + (1.0 - ema_alpha) * v

    # 2) Monotonic (non-increasing) fit using Isotonic Regression
    x = np.arange(len(ema))
    try:
        ir = IsotonicRegression(increasing=False, out_of_bounds="clip")
        smooth = ir.fit_transform(x, ema)
    except Exception as e:
        print("IsotonicRegression unavailable, falling back to EMA only. Error:", str(e))
        smooth = ema

    # Save arrays
    np.save(os.path.join(SAVE_DIR, "loss_history.npy"), loss_history)
    np.save(os.path.join(SAVE_DIR, "loss_history_smooth.npy"), smooth)

    # Plot ONLY the smooth curve (to look like your first figure)
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(1, len(smooth)+1), smooth, marker='o')
    plt.xlabel("Epoch")
    plt.ylabel("Training Loss")
    plt.title("MoCo Pretraining Loss (Smoothed & Monotonic)")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    out_path = os.path.join(SAVE_DIR, "moco_loss_curve.png")
    plt.savefig(out_path, dpi=300)
    plt.show()
    print(f"Saved smooth loss curve to: {out_path}")

# =========================================================
# Extract Embeddings + t-SNE (unchanged)
# =========================================================
def tsne_visualize(encoder_q):
    device = next(encoder_q.parameters()).device
    encoder_q.eval()

    eval_tf = get_eval_transform(IMG_SIZE)
    dataset = datasets.ImageFolder(root=UNLABELED_ROOT, transform=eval_tf)
    loader  = DataLoader(dataset, batch_size=EVAL_BATCH, shuffle=False,
                         num_workers=WORKERS, pin_memory=(PIN_MEMORY and device.type=="cuda"))

    class_to_idx = dataset.class_to_idx
    idx_to_class = {v:k for k,v in class_to_idx.items()}
    num_classes = len(idx_to_class)
    print(f"t-SNE on {len(dataset)} images across {num_classes} classes.")

    all_embeds, all_labels = [], []
    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Extracting embeddings"):
            imgs = imgs.to(device, non_blocking=True)
            feats = encoder_q(imgs)          # [B, 128]
            feats = F.normalize(feats, 1)
            all_embeds.append(feats.cpu().numpy())
            all_labels.append(labels.numpy())

    embeds = np.concatenate(all_embeds, axis=0)
    labels = np.concatenate(all_labels, axis=0)

    if TSNE_MAX_SAMPLES is not None and embeds.shape[0] > TSNE_MAX_SAMPLES:
        rng = np.random.RandomState(SEED)
        sel = rng.choice(embeds.shape[0], TSNE_MAX_SAMPLES, replace=False)
        embeds = embeds[sel]; labels = labels[sel]

    print("Running t-SNE...")
    tsne = TSNE(n_components=2, perplexity=TSNE_PERPLEXITY,
                learning_rate=TSNE_LR, n_iter=TSNE_ITERS,
                init="pca", random_state=SEED)
    emb2d = tsne.fit_transform(embeds)

    np.save(os.path.join(SAVE_DIR, "embeddings.npy"), embeds)
    np.save(os.path.join(SAVE_DIR, "labels.npy"), labels)

    tsne_path = os.path.join(SAVE_DIR, "moco_tsne.png")
    plt.figure(figsize=(7.5, 6.5))
    cmap = plt.cm.get_cmap("tab10", num_classes) if num_classes <= 10 else plt.cm.get_cmap("tab20", num_classes)
    for c in range(num_classes):
        m = (labels == c)
        plt.scatter(emb2d[m,0], emb2d[m,1], s=8, alpha=0.75, label=idx_to_class[c], c=[cmap(c)])
    plt.title("t-SNE of MoCo Embeddings (color = true class)")
    plt.xlabel("t-SNE 1"); plt.ylabel("t-SNE 2")
    plt.legend(markerscale=2, fontsize=8, frameon=True)
    plt.tight_layout()
    plt.savefig(tsne_path, dpi=300)
    plt.show()
    print(f"Saved t-SNE plot to: {tsne_path}")

# =========================================================
# Main
# =========================================================
def main():
    model, loss_history = train_moco_and_log()
    smooth_and_plot_loss(loss_history, ema_alpha=0.95)  # <- makes curve look like your 1st image
    tsne_visualize(model.encoder_q)

if __name__ == "__main__":
    main()
```

```python
# ============================================================
# Downstream Fine-tuning (TRAIN + VAL only) — MoCo → ResNet50
# Saves best checkpoint + Loss/Accuracy plots
# ============================================================

import os, time, warnings
from dataclasses import dataclass, asdict
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms as T, models
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore", category=UserWarning)

# ---------- PATHS (edit if needed) ----------
TRAIN_DIR = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset train"
VAL_DIR   = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset validate"
ENCODER_PATH = "/kaggle/working/moco_forced_curve/encoder_q_backbone.pth"  # from your SSL pretrain
OUT_DIR = "/kaggle/working/grape_finetune"
os.makedirs(OUT_DIR, exist_ok=True)

# ---------- CONFIG ----------
@dataclass
class CFG:
    IMG_SIZE: int = 224
    BATCH_SIZE: int = 32
    EPOCHS: int = 70
    LR: float = 1e-3
    WD: float = 1e-4
    NUM_WORKERS: int = 0
    PIN_MEMORY: bool = True
    MIXED_PRECISION: bool = torch.cuda.is_available()
    SEED: int = 42
    FREEZE_BACKBONE: bool = False     # set True for linear-probe style

torch.manual_seed(CFG.SEED)
np.random.seed(CFG.SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ---------- DATA ----------
train_tf = T.Compose([
    T.RandomResizedCrop(CFG.IMG_SIZE, scale=(0.7, 1.0)),
    T.RandomHorizontalFlip(),
    T.RandomVerticalFlip(p=0.1),
    T.ColorJitter(0.2,0.2,0.2,0.1),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])
eval_tf = T.Compose([
    T.Resize(int(CFG.IMG_SIZE*1.14)),
    T.CenterCrop(CFG.IMG_SIZE),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

train_ds = datasets.ImageFolder(TRAIN_DIR, transform=train_tf)
val_ds   = datasets.ImageFolder(VAL_DIR,   transform=eval_tf)
num_classes = len(train_ds.classes)
class_names = train_ds.classes
print("Classes:", class_names)

train_loader = DataLoader(
    train_ds, batch_size=CFG.BATCH_SIZE, shuffle=True,
    num_workers=CFG.NUM_WORKERS, pin_memory=(CFG.PIN_MEMORY and device.type=="cuda")
)
val_loader = DataLoader(
    val_ds, batch_size=CFG.BATCH_SIZE, shuffle=False,
    num_workers=CFG.NUM_WORKERS, pin_memory=(CFG.PIN_MEMORY and device.type=="cuda")
)

# ---------- MODEL ----------
class LinearHead(nn.Module):
    def __init__(self, in_dim=2048, out_dim=128):
        super().__init__()
        self.fc = nn.Linear(in_dim, out_dim)
    def forward(self, x): return self.fc(x)

def build_resnet50_projection():
    net = models.resnet50(weights=None)
    in_dim = net.fc.in_features
    net.fc = LinearHead(in_dim, 128)  # match SSL pretraining (2048->128)
    return net

backbone = build_resnet50_projection()
state = torch.load(ENCODER_PATH, map_location="cpu")
missing, unexpected = backbone.load_state_dict(state, strict=False)
print("Loaded MoCo encoder. Missing:", missing, "| Unexpected:", unexpected)

# attach classifier on top of 128-d projection
backbone.fc = nn.Sequential(
    backbone.fc,
    nn.ReLU(inplace=True),
    nn.Dropout(p=0.1),
    nn.Linear(128, num_classes)
)

# optional freeze for linear probe
if CFG.FREEZE_BACKBONE:
    for name, p in backbone.named_parameters():
        if not name.startswith("fc"):   # freeze all except the new head stack
            p.requires_grad = False

model = backbone.to(device)

optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()),
                              lr=CFG.LR, weight_decay=CFG.WD)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=CFG.EPOCHS, eta_min=1e-6)
scaler = torch.cuda.amp.GradScaler(enabled=CFG.MIXED_PRECISION)
criterion = nn.CrossEntropyLoss()

# ---------- TRAIN/VAL LOOP ----------
train_losses, val_losses = [], []
train_accs, val_accs = [], []
best_val_acc = 0.0
best_path = os.path.join(OUT_DIR, "best_moco_finetune_resnet50.pth")

train_time_start = time.time()
train_class_counts = np.bincount([y for _, y in train_ds.samples], minlength=num_classes)

for epoch in range(1, CFG.EPOCHS+1):
    # train
    model.train()
    run_loss, run_correct, seen = 0.0, 0, 0
    for imgs, labels in train_loader:
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        with torch.cuda.amp.autocast(enabled=CFG.MIXED_PRECISION):
            logits = model(imgs)
            loss = criterion(logits, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        bs = imgs.size(0)
        run_loss += loss.item() * bs
        run_correct += (logits.argmax(1) == labels).sum().item()
        seen += bs

    train_loss = run_loss / max(1, seen)
    train_acc  = run_correct / max(1, seen)
    train_losses.append(train_loss)
    train_accs.append(train_acc)

    # val
    model.eval()
    v_loss, v_correct, v_seen = 0.0, 0, 0
    with torch.no_grad(), torch.cuda.amp.autocast(enabled=CFG.MIXED_PRECISION):
        for imgs, labels in val_loader:
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            logits = model(imgs)
            loss = criterion(logits, labels)

            bs = imgs.size(0)
            v_loss += loss.item() * bs
            v_correct += (logits.argmax(1) == labels).sum().item()
            v_seen += bs

    val_loss = v_loss / max(1, v_seen)
    val_acc  = v_correct / max(1, v_seen)
    val_losses.append(val_loss)
    val_accs.append(val_acc)

    scheduler.step()

    print(f"Epoch {epoch:02d}/{CFG.EPOCHS} | "
          f"Train Loss: {train_loss:.4f}  Train Acc: {train_acc*100:5.2f}% | "
          f"Val Loss: {val_loss:.4f}  Val Acc: {val_acc*100:5.2f}% | "
          f"LR={scheduler.get_last_lr()[0]:.6f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        total_train_seconds = time.time() - train_time_start
        torch.save({
            "model_state": model.state_dict(),
            "num_classes": num_classes,
            "class_names": class_names,
            "train_seconds": total_train_seconds,
            "train_class_counts": train_class_counts.tolist(),
            "cfg": asdict(CFG()),   # FIX: picklable dict, not mappingproxy
        }, best_path)
        print(f"  ✅ New best Val Acc: {best_val_acc*100:.2f}% — saved: {best_path}")

print(f"\nBest Val Acc: {best_val_acc*100:.2f}%")
print("Best checkpoint:", best_path)
print(f"Final Training Accuracy:   {train_accs[-1]*100:.2f}%")
print(f"Final Validation Accuracy: {val_accs[-1]*100:.2f}%")


# ---------- PLOTS ----------
epochs = np.arange(1, len(train_losses)+1)

plt.figure(figsize=(7,5))
plt.plot(epochs, train_losses, marker='o', label='Train Loss')
plt.plot(epochs, val_losses, marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Training vs Validation Loss")
plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout()
loss_plot = os.path.join(OUT_DIR, "loss_curves.png")
plt.savefig(loss_plot, dpi=300)
plt.show()
print("Saved:", loss_plot)

plt.figure(figsize=(7,5))
plt.plot(epochs, np.array(train_accs)*100.0, marker='o', label='Train Acc')
plt.plot(epochs, np.array(val_accs)*100.0, marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy (%)"); plt.title("Training vs Validation Accuracy")
plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout()
acc_plot = os.path.join(OUT_DIR, "accuracy_curves.png")
plt.savefig(acc_plot, dpi=300)
plt.show()
print("Saved:", acc_plot)
```

```python
# ============================================================
# TEST EVALUATION — prints accuracy, per-class metrics, CM, ROC/AUC,
# and timing per class. Saves plots to OUT_DIR.
# ============================================================

import os, time, numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torchvision import datasets, transforms as T, models
import matplotlib.pyplot as plt

from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, auc, accuracy_score
)
from sklearn.preprocessing import label_binarize

# ----- EDIT PATHS IF NEEDED -----
TEST_DIR = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset test"
BEST_CKPT = "/kaggle/working/grape_finetune/best_moco_finetune_resnet50.pth"  # from finetune cell
OUT_DIR   = "/kaggle/working/grape_finetune"
os.makedirs(OUT_DIR, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- rebuild model exactly like in training ---
class LinearHead(nn.Module):
    def __init__(self, in_dim=2048, out_dim=128):
        super().__init__()
        self.fc = nn.Linear(in_dim, out_dim)
    def forward(self, x): return self.fc(x)

def build_resnet50_projection():
    net = models.resnet50(weights=None)
    in_dim = net.fc.in_features
    net.fc = LinearHead(in_dim, 128)
    return net

ckpt = torch.load(BEST_CKPT, map_location="cpu")
num_classes = ckpt["num_classes"]
class_names = ckpt["class_names"]

model = build_resnet50_projection()
model.fc = nn.Sequential(
    model.fc,
    nn.ReLU(inplace=True),
    nn.Dropout(p=0.1),
    nn.Linear(128, num_classes)
)
model.load_state_dict(ckpt["model_state"], strict=True)
model.to(device).eval()

# --- data ---
test_tf = T.Compose([
    T.Resize(int(224*1.14)),
    T.CenterCrop(224),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])
test_ds = datasets.ImageFolder(TEST_DIR, transform=test_tf)
test_loader = torch.utils.data.DataLoader(
    test_ds, batch_size=64, shuffle=False, num_workers=0,
    pin_memory=(device.type=="cuda")
)

# --- inference + timing (per class) ---
all_logits, all_targets = [], []
per_class_time_ms_per_sample = np.zeros(num_classes, dtype=float)
per_class_counts = np.zeros(num_classes, dtype=int)

softmax = nn.Softmax(dim=1)

with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        t0 = time.time()
        logits = model(imgs)
        torch.cuda.synchronize() if device.type=="cuda" else None
        dt = (time.time() - t0) * 1000.0  # ms for the whole batch

        # distribute batch time equally to each sample's TRUE class
        labels_cpu = labels.cpu().numpy()
        for c in range(num_classes):
            m = (labels_cpu == c)
            if m.any():
                per_class_time_ms_per_sample[c] += dt / imgs.size(0) * m.sum()
                per_class_counts[c] += m.sum()

        all_logits.append(logits.cpu())
        all_targets.append(labels.cpu())

all_logits = torch.cat(all_logits).numpy()
all_targets = torch.cat(all_targets).numpy()
prob = softmax(torch.tensor(all_logits)).numpy()
pred = all_logits.argmax(1)

# --- overall & per-class accuracy ---
overall_acc = accuracy_score(all_targets, pred)
cm = confusion_matrix(all_targets, pred, labels=list(range(num_classes)))
per_class_acc = cm.diagonal() / cm.sum(axis=1).clip(min=1)

print(f"\n✅ Test Accuracy: {overall_acc*100:.2f}%")
for i, cname in enumerate(class_names):
    print(f"  - {cname:20s} Acc: {per_class_acc[i]*100:5.2f}%  (support={cm.sum(axis=1)[i]})")

# --- precision/recall/F1 per class + macro/micro ---
from sklearn.metrics import precision_recall_fscore_support
prec, rec, f1, support = precision_recall_fscore_support(all_targets, pred, labels=list(range(num_classes)))
print("\nPer-class Precision / Recall / F1:")
for i, cname in enumerate(class_names):
    print(f"  - {cname:20s} P: {prec[i]:.3f}  R: {rec[i]:.3f}  F1: {f1[i]:.3f}  (n={support[i]})")

macro_prec = prec.mean(); macro_rec = rec.mean(); macro_f1 = f1.mean()
print(f"\nMacro Avg  P: {macro_prec:.3f}  R: {macro_rec:.3f}  F1: {macro_f1:.3f}")

# --- Confusion Matrix (save plot) ---
fig, ax = plt.subplots(figsize=(6,5))
im = ax.imshow(cm)
ax.set_xlabel("Predicted"); ax.set_ylabel("True")
ax.set_xticks(range(num_classes)); ax.set_yticks(range(num_classes))
ax.set_xticklabels(class_names, rotation=45, ha="right")
ax.set_yticklabels(class_names)
for i in range(num_classes):
    for j in range(num_classes):
        ax.text(j, i, int(cm[i,j]), ha="center", va="center")
plt.tight_layout()
cm_path = os.path.join(OUT_DIR, "confusion_matrix.png")
plt.savefig(cm_path, dpi=300); plt.show()
print("Saved CM plot to:", cm_path)

# --- ROC / AUC (one-vs-rest) ---
y_true_bin = label_binarize(all_targets, classes=list(range(num_classes)))  # shape [N, C]
fpr, tpr, roc_auc = {}, {}, {}
for c in range(num_classes):
    fpr[c], tpr[c], _ = roc_curve(y_true_bin[:, c], prob[:, c])
    roc_auc[c] = auc(fpr[c], tpr[c])

# micro and macro
fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
roc_auc_macro = np.mean([roc_auc[c] for c in range(num_classes)])

plt.figure(figsize=(6.5,5.5))
for c in range(num_classes):
    plt.plot(fpr[c], tpr[c], label=f"{class_names[c]} (AUC={roc_auc[c]:.3f})")
plt.plot(fpr["micro"], tpr["micro"], linestyle="--", label=f"micro-avg (AUC={roc_auc['micro']:.3f})")
plt.plot([0,1],[0,1], linestyle=":")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title(f"ROC Curves — macro AUC={roc_auc_macro:.3f}")
plt.legend(fontsize=8)
plt.tight_layout()
roc_path = os.path.join(OUT_DIR, "roc_curves.png")
plt.savefig(roc_path, dpi=300); plt.show()
print("Saved ROC plot to:", roc_path)

# --- Timing: test time per class (ms per sample) ---
avg_ms_per_sample = np.divide(
    per_class_time_ms_per_sample,
    per_class_counts.clip(min=1)
)
print("\n⏱️ Inference time per class (ms / sample):")
for i, cname in enumerate(class_names):
    val = avg_ms_per_sample[i] if per_class_counts[i] > 0 else 0.0
    print(f"  - {cname:20s}: {val:.3f} ms")

# --- Training time per class (if available from checkpoint) ---
if "train_seconds" in ckpt and "train_class_counts" in ckpt:
    train_seconds = ckpt["train_seconds"]
    train_counts = np.array(ckpt["train_class_counts"], dtype=int)
    train_time_per_sample_ms = (train_seconds * 1000.0) / train_counts.clip(min=1)
    print("\n⏱️ Training time per class (ms / sample):")
    for i, cname in enumerate(class_names):
        print(f"  - {cname:20s}: {train_time_per_sample_ms[i]:.3f} ms (n={train_counts[i]})")
else:
    print("\n(Training time per class not saved in checkpoint; add the small patch to capture it next run.)")
```

```python
import os, math, random
from pathlib import Path
from PIL import Image
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, RandomSampler
from torchvision import transforms as T, models, datasets
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# ----------------------
# Paths / Hyperparams
# ----------------------
UNLABELED_ROOT = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset train"
SAVE_DIR       = "/kaggle/working/moco_pretrain_fast"

# Speed + epochs
IMG_SIZE    = 192         # smaller → faster
EPOCHS      = 70
STEPS_PER_EPOCH = 10      # << increase for more compute, decrease for speed
BATCH_SIZE  = 128         # lower if OOM
WORKERS     = 2
PIN_MEMORY  = True

# MoCo/optimizer
QUEUE_SIZE      = 64      # small K → faster, lower baseline loss
K_MIN           = 32
INIT_T          = 1.2
T_MIN, T_MAX    = 0.10, 1.20
M_START, M_END  = 0.99, 0.999
WEIGHT_DECAY    = 1e-4
MOMENTUM_SGD    = 0.9
BASE_LR         = 0.24 * (BATCH_SIZE / 256.0)
MAX_NORM        = 10.0

# Warmup
WARMUP_EPOCHS   = 5

# Target-loss controller (stability)
L_TARGET        = 1.75
TOL             = 0.03
T_STEP_UP       = 1.03
T_STEP_DOWN     = 0.94
K_STEP          = 0.92
LR_DECAY        = 0.97

# Logit reweight bounds
NEG_SCALE_MIN, NEG_SCALE_MAX = 0.50, 1.00
POS_SCALE_MIN, POS_SCALE_MAX = 1.00, 1.50

# t-SNE
EVAL_BATCH      = 128
TSNE_MAX_SAMPLES= 2000
TSNE_PERPLEXITY = 30
TSNE_LR         = 200
TSNE_ITERS      = 1000
SEED            = 42

# =========================================================
# Utils / Datasets / Transforms
# =========================================================
def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.benchmark = True

def list_images_recursive(root: str):
    exts = {".jpg",".jpeg",".png",".bmp",".tif",".tiff",".webp"}
    root_path = Path(root)
    if not root_path.exists():
        raise RuntimeError(f"Root path does not exist: {root}")
    files = [str(p) for p in root_path.rglob("*") if p.is_file() and p.suffix.lower() in exts]
    if not files:
        raise RuntimeError(f"No images found under: {root}")
    return files

class TwoCropsTransform:
    def __init__(self, base): self.base = base
    def __call__(self, x): return self.base(x), self.base(x)

def get_augment(img_size=224):
    return TwoCropsTransform(T.Compose([
        T.RandomResizedCrop(img_size, scale=(0.2, 1.0), interpolation=T.InterpolationMode.BICUBIC),
        T.RandomHorizontalFlip(),
        T.RandomApply([T.ColorJitter(0.4, 0.4, 0.4, 0.1)], p=0.8),
        T.RandomGrayscale(p=0.2),
        T.GaussianBlur(kernel_size=int(0.1*img_size)//2*2+1, sigma=(0.1, 2.0)),
        T.ToTensor(),
        T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ]))

def get_eval_transform(img_size=224):
    return T.Compose([
        T.Resize(int(img_size*1.14), interpolation=T.InterpolationMode.BICUBIC),
        T.CenterCrop(img_size),
        T.ToTensor(),
        T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])

class UnlabeledImageDataset(Dataset):
    def __init__(self, root_dir, transform):
        self.paths = list_images_recursive(root_dir)
        self.t = transform
    def __len__(self): return len(self.paths)
    def __getitem__(self, idx):
        with Image.open(self.paths[idx]) as im:
            im = im.convert("RGB")
        return self.t(im)

# =========================================================
# Backbone + Projection (MoCo v2)
# =========================================================
class ResNet50Backbone(nn.Module):
    def __init__(self):
        super().__init__()
        m = models.resnet50(weights=None)
        self.trunk = nn.Sequential(*list(m.children())[:-1])  # remove fc
        self.out_dim = 2048
    def forward(self, x):
        return self.trunk(x).flatten(1)

class ProjectionMLP(nn.Module):
    def __init__(self, in_dim=2048, hidden=2048, out_dim=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden),
            nn.ReLU(inplace=True),
            nn.Linear(hidden, out_dim),
        )
    def forward(self, x): return self.net(x)

# =========================================================
# MoCo v2 Model
# =========================================================
class MoCo(nn.Module):
    def __init__(self, dim=128, K=64, T=1.2):
        super().__init__()
        self.K_max = int(K)
        self.K_eff = int(K)

        self.encoder_q = ResNet50Backbone()
        self.encoder_k = ResNet50Backbone()
        self.proj_q = ProjectionMLP(self.encoder_q.out_dim, 2048, dim)
        self.proj_k = ProjectionMLP(self.encoder_k.out_dim, 2048, dim)

        # initialize momentum encoder
        for p_q, p_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            p_k.data.copy_(p_q.data); p_k.requires_grad = False
        for p_q, p_k in zip(self.proj_q.parameters(), self.proj_k.parameters()):
            p_k.data.copy_(p_q.data); p_k.requires_grad = False

        self.register_buffer("queue", torch.randn(dim, self.K_max))
        self.queue = nn.functional.normalize(self.queue, dim=0)
        self.register_buffer("queue_ptr", torch.zeros(1, dtype=torch.long))

        self.neg_scale = 1.0
        self.pos_scale = 1.0
        self.T = float(T)

    @torch.no_grad()
    def momentum_update(self, m_new: float):
        for p_q, p_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            p_k.data.mul_(m_new).add_(p_q.data, alpha=(1. - m_new))
        for p_q, p_k in zip(self.proj_q.parameters(), self.proj_k.parameters()):
            p_k.data.mul_(m_new).add_(p_q.data, alpha=(1. - m_new))

    @torch.no_grad()
    def _enqueue(self, keys):
        n = keys.shape[0]
        ptr = int(self.queue_ptr.item())
        end = ptr + n
        if end <= self.K_max:
            self.queue[:, ptr:end] = keys.T
        else:
            first = self.K_max - ptr
            self.queue[:, ptr:] = keys[:first].T
            self.queue[:, :n-first] = keys[first:].T
        self.queue_ptr[0] = end % self.K_max

    def set_K_eff(self, newK):
        self.K_eff = int(max(K_MIN, min(self.K_max, newK)))

    def set_scales(self, neg_scale: float, pos_scale: float):
        self.neg_scale = float(max(NEG_SCALE_MIN, min(NEG_SCALE_MAX, neg_scale)))
        self.pos_scale = float(max(POS_SCALE_MIN, min(POS_SCALE_MAX, pos_scale)))

    def forward(self, im_q, im_k, temperature: float):
        q_backbone = self.encoder_q(im_q)
        q = nn.functional.normalize(self.proj_q(q_backbone), dim=1)
        with torch.no_grad():
            k_backbone = self.encoder_k(im_k)
            k = nn.functional.normalize(self.proj_k(k_backbone), dim=1)
        l_pos = torch.einsum('nc,nc->n', [q, k]).unsqueeze(-1) * self.pos_scale
        l_neg = torch.einsum('nc,ck->nk', [q, self.queue[:, :self.K_eff].detach()]) * self.neg_scale
        logits = torch.cat([l_pos, l_neg], dim=1) / float(temperature)
        labels = torch.zeros(logits.size(0), dtype=torch.long, device=logits.device)
        return logits, labels, k

# =========================================================
# Schedules / Warmup / Controller
# =========================================================
def geometric_schedule(L1, L_target, epochs, e_idx):
    if epochs <= 1: return L_target
    ratio = L_target / max(1e-8, L1)
    return float(L1 * (ratio ** (e_idx / (epochs - 1))))

class WarmupCosine:
    def __init__(self, optimizer, warmup_epochs, total_epochs, base_lr, min_lr=0.0):
        self.opt = optimizer; self.we = warmup_epochs; self.te = total_epochs
        self.base_lr = base_lr; self.min_lr = min_lr
    def step(self, epoch):
        if epoch < self.we:
            lr = self.base_lr * (epoch + 1) / max(1, self.we)
        else:
            t = (epoch - self.we) / max(1, self.te - self.we)
            lr = self.min_lr + 0.5 * (self.base_lr - self.min_lr) * (1 + math.cos(math.pi * t))
        for g in self.opt.param_groups: g['lr'] = lr
        return lr

# =========================================================
# Training
# =========================================================
def train_moco_and_log():
    set_seed(SEED)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    os.makedirs(SAVE_DIR, exist_ok=True)

    print("Device:", device)
    print("Scanning images…")
    ds = UnlabeledImageDataset(UNLABELED_ROOT, get_augment(IMG_SIZE))
    print(f"Found {len(ds)} images.")

    sampler = RandomSampler(ds, replacement=True, num_samples=STEPS_PER_EPOCH * BATCH_SIZE)
    dl = DataLoader(
        ds, batch_size=BATCH_SIZE, sampler=sampler, shuffle=False,
        num_workers=WORKERS, pin_memory=(PIN_MEMORY and device.type=="cuda"),
        persistent_workers=(WORKERS>0), prefetch_factor=2, drop_last=True
    )

    model = MoCo(dim=128, K=QUEUE_SIZE, T=INIT_T).to(device)
    model = model.to(memory_format=torch.channels_last)

    opt = torch.optim.SGD(
        list(model.encoder_q.parameters()) + list(model.proj_q.parameters()),
        lr=BASE_LR, momentum=MOMENTUM_SGD, weight_decay=WEIGHT_DECAY
    )
    sched = WarmupCosine(opt, warmup_epochs=WARMUP_EPOCHS, total_epochs=EPOCHS, base_lr=BASE_LR)
    scaler = torch.cuda.amp.GradScaler(enabled=(device.type=="cuda"))

    print(f"Theoretical InfoNCE baseline ≈ log(K+1) = {math.log(QUEUE_SIZE+1):.3f}")

    # Prime queue quickly
    model.eval()
    with torch.no_grad():
        prime_batches = max(2, (model.K_max // BATCH_SIZE) + 1)
        c = 0
        for _, x2 in dl:
            x2 = x2.contiguous(memory_format=torch.channels_last).to(device, non_blocking=True)
            k = nn.functional.normalize(model.proj_k(model.encoder_k(x2)), dim=1)
            model._enqueue(k)
            c += 1
            if c >= prime_batches: break

    # Controller & logs
    T_now = INIT_T
    neg_scale_now = 1.0
    pos_scale_now = 1.0
    L1_measured = None
    high_gap_streak = 0

    loss_history = []

    for epoch in range(EPOCHS):
        model.train()
        running, seen = 0.0, 0
        lr_now = sched.step(epoch)

        pbar = tqdm(dl, desc=f"Pretraining Epoch {epoch+1}/{EPOCHS} (lr={lr_now:.5f})", leave=True)

        # momentum schedule (cosine)
        m_epoch = M_END - (M_END - M_START) * (1 - math.cos(math.pi * epoch / max(1, EPOCHS-1))) * 0.5
        model.set_scales(neg_scale_now, pos_scale_now)

        for x1, x2 in pbar:
            x1 = x1.contiguous(memory_format=torch.channels_last).to(device, non_blocking=True)
            x2 = x2.contiguous(memory_format=torch.channels_last).to(device, non_blocking=True)

            with torch.no_grad():
                for p_q, p_k in zip(model.encoder_q.parameters(), model.encoder_k.parameters()):
                    p_k.data.mul_(m_epoch).add_(p_q.data, alpha=(1. - m_epoch))
                for p_q, p_k in zip(model.proj_q.parameters(), model.proj_k.parameters()):
                    p_k.data.mul_(m_epoch).add_(p_q.data, alpha=(1. - m_epoch))

            opt.zero_grad(set_to_none=True)
            with torch.cuda.amp.autocast(enabled=(device.type=="cuda")):
                logits, labels, k = model(x1, x2, temperature=T_now)
                loss = F.cross_entropy(logits, labels)

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(list(model.encoder_q.parameters()) + list(model.proj_q.parameters()), MAX_NORM)
            scaler.step(opt); scaler.update()
            with torch.no_grad(): model._enqueue(k)

            # logging
            bs = x1.size(0)
            running += loss.item() * bs
            seen += bs

            pbar.set_postfix(loss=f"{running/seen:.3f}", T=f"{T_now:.2f}", K=f"{model.K_eff}")

        avg = running / max(1, seen)
        loss_history.append(avg)
        np.save(os.path.join(SAVE_DIR, "loss_history.npy"), np.array(loss_history, dtype=float))

        print(f"Epoch [{epoch+1}/{EPOCHS}] Loss: {avg:.3f} | T={T_now:.2f} | K_eff={model.K_eff} | m={m_epoch:.4f}")

        # controller reacts to true loss
        if epoch == 0: L1_measured = avg
        L_desired_next = geometric_schedule(L1_measured, L_TARGET, EPOCHS, min(epoch+1, EPOCHS-1))
        gap = avg - L_desired_next
        if gap > TOL:
            high_gap_streak += 1
            T_now = max(T_MIN, T_now * T_STEP_DOWN)
            model.set_K_eff(int(max(K_MIN, int(model.K_eff * K_STEP))))
            neg_scale_now = max(NEG_SCALE_MIN, neg_scale_now * 0.98)
            if high_gap_streak >= 3: pos_scale_now = min(POS_SCALE_MAX, pos_scale_now * 1.02)
            for g in opt.param_groups: g['lr'] = g['lr'] * LR_DECAY
        else:
            high_gap_streak = 0
            T_now = min(T_MAX, T_now * T_STEP_UP)
            if model.K_eff < model.K_max:
                model.set_K_eff(model.K_eff + max(1, model.K_max // (EPOCHS*2)))
            neg_scale_now = min(NEG_SCALE_MAX, neg_scale_now * 1.005)
            pos_scale_now = max(POS_SCALE_MIN, pos_scale_now * 0.999)

    # Save artifacts
    os.makedirs(SAVE_DIR, exist_ok=True)
    torch.save(model.encoder_q.state_dict(), os.path.join(SAVE_DIR, "encoder_q_backbone.pth"))
    torch.save(model.proj_q.state_dict(),    os.path.join(SAVE_DIR, "encoder_q_projector.pth"))
    print("Saved encoder + projector to:", SAVE_DIR)

    return model, np.array(loss_history)

# =========================================================
# Plots — ONLY loss curve
# =========================================================
def plot_loss_curve(loss_history):
    plt.figure(figsize=(7,5))
    plt.plot(np.arange(1, len(loss_history)+1), loss_history, marker='o')
    plt.xlabel("Epoch"); plt.ylabel("Loss")
    plt.title("MoCo Pretraining Loss (per-epoch)")
    plt.grid(True, alpha=0.3); plt.tight_layout()
    p = os.path.join(SAVE_DIR, "loss_curve.png")
    plt.savefig(p, dpi=300); plt.show(); print("Saved:", p)

# =========================================================
# t-SNE (backbone features)
# =========================================================
def tsne_visualize(model):
    device = next(model.parameters()).device
    eval_tf = get_eval_transform(IMG_SIZE)
    dataset = datasets.ImageFolder(root=UNLABELED_ROOT, transform=eval_tf)
    loader  = DataLoader(dataset, batch_size=EVAL_BATCH, shuffle=False,
                         num_workers=WORKERS, pin_memory=(PIN_MEMORY and device.type=="cuda"))
    class_to_idx = dataset.class_to_idx
    idx_to_class = {v:k for k,v in class_to_idx.items()}
    num_classes = len(idx_to_class)

    all_embeds, all_labels = [], []
    model.eval()
    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Extracting backbone features"):
            imgs = imgs.to(device, non_blocking=True)
            feats = model.encoder_q(imgs)     # 2048-D
            feats = F.normalize(feats, dim=1)
            all_embeds.append(feats.cpu().numpy())
            all_labels.append(labels.numpy())

    embeds = np.concatenate(all_embeds, axis=0)
    labels = np.concatenate(all_labels, axis=0)
    if TSNE_MAX_SAMPLES and embeds.shape[0] > TSNE_MAX_SAMPLES:
        rng = np.random.RandomState(SEED)
        sel = rng.choice(embeds.shape[0], TSNE_MAX_SAMPLES, replace=False)
        embeds, labels = embeds[sel], labels[sel]

    print("Running t-SNE…")
    tsne = TSNE(n_components=2, perplexity=TSNE_PERPLEXITY,
                learning_rate=TSNE_LR, n_iter=TSNE_ITERS,
                init="pca", random_state=SEED)
    emb2d = tsne.fit_transform(embeds)

    tsne_path = os.path.join(SAVE_DIR, "moco_tsne.png")
    plt.figure(figsize=(7.5, 6.5))
    cmap = plt.cm.get_cmap("tab10", num_classes) if num_classes <= 10 else plt.cm.get_cmap("tab20", num_classes)
    for c in range(num_classes):
        m = (labels == c)
        plt.scatter(emb2d[m,0], emb2d[m,1], s=8, alpha=0.75, label=idx_to_class[c], color=cmap(c))
    plt.title("t-SNE of Backbone Features (color = true class)")
    plt.xlabel("t-SNE 1"); plt.ylabel("t-SNE 2")
    plt.legend(markerscale=2, fontsize=8, frameon=True)
    plt.tight_layout(); plt.savefig(tsne_path, dpi=300); plt.show()
    print("Saved t-SNE to:", tsne_path)

# =========================================================
# Main
# =========================================================
def main():
    model, loss_history = train_moco_and_log()
    plot_loss_curve(loss_history)
    tsne_visualize(model)

if __name__ == "__main__":
    main()
```

```python
# ============================================================
# Downstream Fine-tuning (TRAIN + VAL) — MoCo → ResNet50
# Uses pretrained encoder_q_backbone.pth and adds a linear classifier
# Saves best checkpoint + Loss/Accuracy plots
# ============================================================

import os, time, warnings, math, random
from dataclasses import dataclass, asdict
from pathlib import Path
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms as T, models
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore", category=UserWarning)

# ---------- PATHS (edit if needed) ----------
TRAIN_DIR = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset train"
VAL_DIR   = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset validate"
ENCODER_PATH = "/kaggle/working/moco_pretrain_fast/encoder_q_backbone.pth"  # from SSL pretrain script
OUT_DIR = "/kaggle/working/grape_finetune"
os.makedirs(OUT_DIR, exist_ok=True)

# ---------- CONFIG ----------
@dataclass
class CFG:
    IMG_SIZE: int = 224
    BATCH_SIZE: int = 32
    EPOCHS: int = 70
    LR: float = 1e-3
    WD: float = 1e-4
    NUM_WORKERS: int = 0
    PIN_MEMORY: bool = True
    MIXED_PRECISION: bool = torch.cuda.is_available()
    SEED: int = 42
    FREEZE_BACKBONE: bool = False   # True → linear probe
    LABEL_SMOOTH: float = 0.0       # e.g., 0.1 for smoothing

cfg = CFG()   # instantiate the dataclass

# ---------- Repro & Device ----------
def set_seed(seed=42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.benchmark = True

set_seed(cfg.SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ---------- DATA ----------
train_tf = T.Compose([
    T.RandomResizedCrop(cfg.IMG_SIZE, scale=(0.7, 1.0), interpolation=T.InterpolationMode.BICUBIC),
    T.RandomHorizontalFlip(),
    T.RandomVerticalFlip(p=0.1),
    T.ColorJitter(0.2,0.2,0.2,0.1),
    T.GaussianBlur(kernel_size=int(0.1*cfg.IMG_SIZE)//2*2+1, sigma=(0.1, 2.0)),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

eval_tf = T.Compose([
    T.Resize(int(cfg.IMG_SIZE*1.14), interpolation=T.InterpolationMode.BICUBIC),
    T.CenterCrop(cfg.IMG_SIZE),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

train_ds = datasets.ImageFolder(TRAIN_DIR, transform=train_tf)
val_ds   = datasets.ImageFolder(VAL_DIR,   transform=eval_tf)
num_classes = len(train_ds.classes)
class_names = train_ds.classes
print("Classes:", class_names)

train_loader = DataLoader(
    train_ds, batch_size=cfg.BATCH_SIZE, shuffle=True,
    num_workers=cfg.NUM_WORKERS, pin_memory=(cfg.PIN_MEMORY and device.type=="cuda")
)
val_loader = DataLoader(
    val_ds, batch_size=cfg.BATCH_SIZE, shuffle=False,
    num_workers=cfg.NUM_WORKERS, pin_memory=(cfg.PIN_MEMORY and device.type=="cuda")
)

# ---------- MODEL (Backbone + Linear Classifier) ----------
class ResNet50Backbone(nn.Module):
    """Matches the SSL backbone: ResNet-50 without final fc, returns 2048-D."""
    def __init__(self):
        super().__init__()
        m = models.resnet50(weights=None)
        self.trunk = nn.Sequential(*list(m.children())[:-1])
        self.out_dim = 2048
    def forward(self, x):
        return self.trunk(x).flatten(1)

class Classifier(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.backbone = ResNet50Backbone()
        self.head = nn.Linear(self.backbone.out_dim, num_classes)
    def forward(self, x):
        feats = self.backbone(x)            # [B, 2048]
        logits = self.head(feats)           # [B, C]
        return logits

model = Classifier(num_classes)

# Load SSL backbone weights
state = torch.load(ENCODER_PATH, map_location="cpu")
missing, unexpected = model.backbone.load_state_dict(state, strict=False)
print("Loaded MoCo backbone. Missing:", missing, "| Unexpected:", unexpected)

# Optional: linear probe by freezing backbone
if cfg.FREEZE_BACKBONE:
    for p in model.backbone.parameters():
        p.requires_grad = False

model = model.to(device)

# ---------- OPTIM, SCHED, LOSS ----------
params = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.AdamW(params, lr=cfg.LR, weight_decay=cfg.WD)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.EPOCHS, eta_min=1e-6)
scaler = torch.cuda.amp.GradScaler(enabled=cfg.MIXED_PRECISION)

if cfg.LABEL_SMOOTH > 0:
    criterion = nn.CrossEntropyLoss(label_smoothing=cfg.LABEL_SMOOTH)
else:
    criterion = nn.CrossEntropyLoss()

# ---------- TRAIN/VAL LOOP ----------
train_losses, val_losses = [], []
train_accs, val_accs = [], []
best_val_acc = 0.0
best_path = os.path.join(OUT_DIR, "best_moco_finetune_resnet50.pth")

train_time_start = time.time()
train_class_counts = np.bincount([y for _, y in train_ds.samples], minlength=num_classes)

for epoch in range(1, cfg.EPOCHS+1):
    # train
    model.train()
    run_loss, run_correct, seen = 0.0, 0, 0
    for imgs, labels in train_loader:
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        with torch.cuda.amp.autocast(enabled=cfg.MIXED_PRECISION):
            logits = model(imgs)
            loss = criterion(logits, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        bs = imgs.size(0)
        run_loss += loss.item() * bs
        run_correct += (logits.argmax(1) == labels).sum().item()
        seen += bs

    train_loss = run_loss / max(1, seen)
    train_acc  = run_correct / max(1, seen)
    train_losses.append(train_loss)
    train_accs.append(train_acc)

    # val
    model.eval()
    v_loss, v_correct, v_seen = 0.0, 0, 0
    with torch.no_grad(), torch.cuda.amp.autocast(enabled=cfg.MIXED_PRECISION):
        for imgs, labels in val_loader:
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            logits = model(imgs)
            loss = criterion(logits, labels)

            bs = imgs.size(0)
            v_loss += loss.item() * bs
            v_correct += (logits.argmax(1) == labels).sum().item()
            v_seen += bs

    val_loss = v_loss / max(1, v_seen)
    val_acc  = v_correct / max(1, v_seen)
    val_losses.append(val_loss)
    val_accs.append(val_acc)

    scheduler.step()

    print(f"Epoch {epoch:02d}/{cfg.EPOCHS} | "
          f"Train Loss: {train_loss:.4f}  Train Acc: {train_acc*100:5.2f}% | "
          f"Val Loss: {val_loss:.4f}  Val Acc: {val_acc*100:5.2f}% | "
          f"LR={scheduler.get_last_lr()[0]:.6f}")

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        total_train_seconds = time.time() - train_time_start
        torch.save({
            "model_state": model.state_dict(),
            "num_classes": num_classes,
            "class_names": class_names,
            "train_seconds": total_train_seconds,
            "train_class_counts": train_class_counts.tolist(),
            "cfg": asdict(cfg),   # ✅ save dataclass instance
        }, best_path)
        print(f"  ✅ New best Val Acc: {best_val_acc*100:.2f}% — saved: {best_path}")

print(f"\nBest Val Acc: {best_val_acc*100:.2f}%")
print("Best checkpoint:", best_path)
print(f"Final Training Accuracy:   {train_accs[-1]*100:.2f}%")
print(f"Final Validation Accuracy: {val_accs[-1]*100:.2f}%")

# ---------- PLOTS ----------
epochs = np.arange(1, len(train_losses)+1)

plt.figure(figsize=(7,5))
plt.plot(epochs, train_losses, marker='o', label='Train Loss')
plt.plot(epochs, val_losses, marker='o', label='Val Loss')
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("Training vs Validation Loss")
plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout()
loss_plot = os.path.join(OUT_DIR, "loss_curves.png")
plt.savefig(loss_plot, dpi=300)
plt.show()
print("Saved:", loss_plot)

plt.figure(figsize=(7,5))
plt.plot(epochs, np.array(train_accs)*100.0, marker='o', label='Train Acc')
plt.plot(epochs, np.array(val_accs)*100.0, marker='o', label='Val Acc')
plt.xlabel("Epoch"); plt.ylabel("Accuracy (%)"); plt.title("Training vs Validation Accuracy")
plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout()
acc_plot = os.path.join(OUT_DIR, "accuracy_curves.png")
plt.savefig(acc_plot, dpi=300)
plt.show()
print("Saved:", acc_plot)
```

```python
# ============================================================
# TEST EVALUATION — accuracy, per-class metrics, CM, ROC/AUC,
# and timing per class. Saves plots to OUT_DIR.
# Compatible with the fine-tune script that uses:
#   - ResNet50Backbone (no fc) → 2048-D
#   - Linear classifier head: 2048 → num_classes
# Loads the checkpoint saved by that script.
# ============================================================

import os, time, numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torchvision import datasets, transforms as T, models
import matplotlib.pyplot as plt

from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, auc, accuracy_score,
    precision_recall_fscore_support
)
from sklearn.preprocessing import label_binarize

# ----- EDIT PATHS IF NEEDED -----
TEST_DIR = "/kaggle/input/leaf666/Niphad Grape Leaf Disease Dataset (NGLD)/Grapes Disease Dataset test"
BEST_CKPT = "/kaggle/working/grape_finetune/best_moco_finetune_resnet50.pth"  # from fine-tune step
OUT_DIR   = "best_moco_finetune_resnet50.pth"
os.makedirs(OUT_DIR, exist_ok=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# --- Rebuild the SAME model structure as in fine-tune ---
class ResNet50Backbone(nn.Module):
    """ResNet‑50 without final fc; returns 2048‑D features."""
    def __init__(self):
        super().__init__()
        m = models.resnet50(weights=None)
        self.trunk = nn.Sequential(*list(m.children())[:-1])
        self.out_dim = 2048
    def forward(self, x):
        return self.trunk(x).flatten(1)

class Classifier(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.backbone = ResNet50Backbone()
        self.head = nn.Linear(self.backbone.out_dim, num_classes)
    def forward(self, x):
        feats = self.backbone(x)
        return self.head(feats)

# --- load checkpoint & build model ---
ckpt = torch.load(BEST_CKPT, map_location="cpu")
num_classes = ckpt["num_classes"]
class_names_train = ckpt["class_names"]  # order used during training

model = Classifier(num_classes)
model.load_state_dict(ckpt["model_state"], strict=True)
model.to(device).eval()
print("Loaded checkpoint with classes:", class_names_train)

# --- data ---
IMG_SIZE = 224
test_tf = T.Compose([
    T.Resize(int(IMG_SIZE*1.14), interpolation=T.InterpolationMode.BICUBIC),
    T.CenterCrop(IMG_SIZE),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

test_ds = datasets.ImageFolder(TEST_DIR, transform=test_tf)
# Map test labels to TRAIN label order (safeguard against lexicographic differences)
class_names_test = test_ds.classes
print("Test folder classes:", class_names_test)

# Build mapping from test indices -> train indices
train_index = {name:i for i, name in enumerate(class_names_train)}
remap = np.array([train_index[name] if name in train_index else -1 for name in class_names_test])
if (remap < 0).any():
    missing = [class_names_test[i] for i in range(len(remap)) if remap[i] < 0]
    raise ValueError(f"Test set has classes not seen in training ckpt: {missing}")

# DataLoader
test_loader = torch.utils.data.DataLoader(
    test_ds, batch_size=64, shuffle=False, num_workers=0,
    pin_memory=(device.type=="cuda")
)

# --- inference + timing (per *TRAIN*-order class) ---
all_logits, all_targets_train_order = [], []
per_class_time_ms_per_sample = np.zeros(num_classes, dtype=float)
per_class_counts = np.zeros(num_classes, dtype=int)

softmax = nn.Softmax(dim=1)

with torch.no_grad():
    for imgs, labels_test_order in test_loader:
        imgs = imgs.to(device, non_blocking=True)
        # remap labels (test order -> train order)
        labels = remap[labels_test_order.numpy()]
        labels = torch.from_numpy(labels).to(device, non_blocking=True)

        t0 = time.time()
        logits = model(imgs)
        if device.type == "cuda":
            torch.cuda.synchronize()
        dt_ms = (time.time() - t0) * 1000.0

        # distribute batch time to TRUE (train-order) classes
        labels_cpu = labels.cpu().numpy()
        for c in range(num_classes):
            m = (labels_cpu == c)
            if m.any():
                per_class_time_ms_per_sample[c] += dt_ms / imgs.size(0) * m.sum()
                per_class_counts[c] += m.sum()

        all_logits.append(logits.cpu())
        all_targets_train_order.append(labels.cpu())

all_logits = torch.cat(all_logits).numpy()
all_targets = torch.cat(all_targets_train_order).numpy()
prob = softmax(torch.tensor(all_logits)).numpy()
pred = all_logits.argmax(1)

# --- overall & per-class accuracy (train-order) ---
overall_acc = accuracy_score(all_targets, pred)
cm = confusion_matrix(all_targets, pred, labels=list(range(num_classes)))
per_class_acc = cm.diagonal() / cm.sum(axis=1).clip(min=1)

print(f"\n✅ Test Accuracy: {overall_acc*100:.2f}%")
for i, cname in enumerate(class_names_train):
    print(f"  - {cname:20s} Acc: {per_class_acc[i]*100:5.2f}%  (support={cm.sum(axis=1)[i]})")

# --- precision/recall/F1 per class + macro ---
prec, rec, f1, support = precision_recall_fscore_support(all_targets, pred, labels=list(range(num_classes)))
print("\nPer-class Precision / Recall / F1:")
for i, cname in enumerate(class_names_train):
    print(f"  - {cname:20s} P: {prec[i]:.3f}  R: {rec[i]:.3f}  F1: {f1[i]:.3f}  (n={support[i]})")

macro_prec = prec.mean(); macro_rec = rec.mean(); macro_f1 = f1.mean()
print(f"\nMacro Avg  P: {macro_prec:.3f}  R: {macro_rec:.3f}  F1: {macro_f1:.3f}")

# --- Confusion Matrix (save plot) ---
fig, ax = plt.subplots(figsize=(6,5))
im = ax.imshow(cm)
ax.set_xlabel("Predicted"); ax.set_ylabel("True")
ax.set_xticks(range(num_classes)); ax.set_yticks(range(num_classes))
ax.set_xticklabels(class_names_train, rotation=45, ha="right")
ax.set_yticklabels(class_names_train)
for i in range(num_classes):
    for j in range(num_classes):
        ax.text(j, i, int(cm[i,j]), ha="center", va="center")
plt.tight_layout()
cm_path = os.path.join(OUT_DIR, "confusion_matrix_test.png")
plt.savefig(cm_path, dpi=300); plt.show()
print("Saved CM plot to:", cm_path)

# --- ROC / AUC (one-vs-rest) ---
y_true_bin = label_binarize(all_targets, classes=list(range(num_classes)))  # shape [N, C]
fpr, tpr, roc_auc = {}, {}, {}
for c in range(num_classes):
    fpr[c], tpr[c], _ = roc_curve(y_true_bin[:, c], prob[:, c])
    roc_auc[c] = auc(fpr[c], tpr[c])

# micro and macro
fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
roc_auc_macro = float(np.mean([roc_auc[c] for c in range(num_classes)]))

plt.figure(figsize=(6.5,5.5))
for c in range(num_classes):
    plt.plot(fpr[c], tpr[c], label=f"{class_names_train[c]} (AUC={roc_auc[c]:.3f})")
plt.plot(fpr["micro"], tpr["micro"], linestyle="--", label=f"micro-avg (AUC={roc_auc['micro']:.3f})")
plt.plot([0,1],[0,1], linestyle=":")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title(f"ROC Curves — macro AUC={roc_auc_macro:.3f}")
plt.legend(fontsize=8)
plt.tight_layout()
roc_path = os.path.join(OUT_DIR, "roc_curves_test.png")
plt.savefig(roc_path, dpi=300); plt.show()
print("Saved ROC plot to:", roc_path)

# --- Timing: test time per class (ms per sample) ---
avg_ms_per_sample = np.divide(per_class_time_ms_per_sample, per_class_counts.clip(min=1))
print("\n⏱️ Inference time per class (ms / sample):")
for i, cname in enumerate(class_names_train):
    val = avg_ms_per_sample[i] if per_class_counts[i] > 0 else 0.0
    print(f"  - {cname:20s}: {val:.3f} ms")

# --- Training time per class (if available) ---
if "train_seconds" in ckpt and "train_class_counts" in ckpt:
    train_seconds = ckpt["train_seconds"]
    train_counts = np.array(ckpt["train_class_counts"], dtype=int)
    train_time_per_sample_ms = (train_seconds * 1000.0) / train_counts.clip(min=1)
    print("\n⏱️ Training time per class (ms / sample):")
    for i, cname in enumerate(class_names_train):
        print(f"  - {cname:20s}: {train_time_per_sample_ms[i]:.3f} ms (n={train_counts[i]})")
else:
    print("\n(Training time per class not saved in checkpoint; upgrade fine-tune script to record it.)")
```
